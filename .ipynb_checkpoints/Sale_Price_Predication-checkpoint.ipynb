# Column Description 


ItemIdentifier ---- Unique product ID

ItemWeight ---- Weight of product

ItemFatContent ---- Whether the product is low fat or not

ItemVisibility ---- The % of the total display area of all products in a store allocated to the particular product

ItemType ---- The category to which the product belongs

ItemMRP ---- Maximum Retail Price (list price) of the product

OutletIdentifier ---- Unique store ID

OutletEstablishmentYear ---- The year in which the store was established

OutletSize ---- The size of the store in terms of ground area covered

OutletLocationType ---- The type of city in which the store is located

OutletType ---- Whether the outlet is just a grocery store or some sort of supermarket

ItemOutletSales ---- sales of the product in t particular store. This is the outcome variable to be predicted.

```python
# Detecting outliers : #An outlier is a data point that lies outside the overall pattern in a distribution . 
# A commonly used rule states that a data point is an outlier if it is more than 1.5*IQR above the 
# third quartile or below the first quartile . 
# Using this , one con remove the outliers and output the resulting data in fill_data variable . 
```

```python
# calculate the first quantile for Item_Visibility
Q1_train=df_train.Item_Visibility.quantile(0.25)
Q1_test=df_test.Item_Visibility.quantile(0.25)

# calculate the second quantile for Item_visibility
Q3_train=df_train.Item_Visibility.quantile(0.75)
Q3_test=df_test.Item_Visibility.quantile(0.75)

print(f" first quantile for train set ={Q1_train}")
print(f" first quantile for test set ={Q1_test}")

print(f" \n second quantile for train set ={Q3_train}")
print(f" second quantile for test set ={Q3_test}")
```

```python
# Calculate the interquartile range ( IQR ) 
IQR_train = Q3_train - Q1_train
IQR_train
```

```python
# Calculate the interquartile range ( IQR ) 
IQR_test = Q3_test - Q1_test
IQR_test
```

```python
# Now that the IQR range is known , remove the outliers from the data 
#The resulting data is stored in new_data variable

new_data_train= df_train.query("(@Q1_train-@IQR_train*1.5) <= Item_Visibility <= (@Q3_train+@IQR_train*1.5)")
new_data_test= df_test.query("(@Q1_test-@IQR_test*1.5) <= Item_Visibility <= (@Q3_test+@IQR_test*1.5)")
```

```python
print(new_data_train.shape)
print(new_data_test.shape)
```

```python
# Shape of the original dataset with outliers is
print(df_train.shape)
print(df_test.shape)
```

```python
# Assiging New_data to orignal_data
df_train=new_data_train
df_test=new_data_test

print(df_train.shape)
print(df_test.shape)
```
